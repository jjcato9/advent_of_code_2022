```python
#Read input file
path = 'input_dec10.txt'

with open(path,mode='r') as file:
    data = file.read()

lines_list = data.split('\n')
lines_list

instructions = []
for line in lines_list:
    instructions.append(line.split(' '))

instructions
```

```python
X = 1
cycle = 1

signal_strength = {}
signal_strength[cycle] = X

for i in range(len(instructions)):
    if instructions[i][0] == 'noop':
        X = X
        cycle += 1
        signal_strength.update({cycle:X})
    if instructions[i][0] == 'addx':
        X = X
        cycle += 1
        signal_strength.update({cycle:X})
        X = X + int(instructions[i][1])
        cycle += 1
        signal_strength.update({cycle:X})
        
signal_strength
```

```python
sum = (20 * signal_strength[20]) +(60 * signal_strength[60]) + (100 * signal_strength[100]) +(140 * signal_strength[140]) +(180 * signal_strength[180]) +(220 * signal_strength[220])
sum
```

```python
output = []
for k, v in signal_strength.items():
    sprite = '...........................................'
    position = v
    new_char = '#'
    temp = list(sprite)
    temp[v] = new_char
    temp[v+1] = new_char
    temp[v+2] = new_char
    sprite = ''.join(temp)
    if k < 41: 
        output.append(sprite[k])
    if k >= 40 and k <80:
        output.append(sprite[k-40])
    if k >= 80 and k <120:
        output.append(sprite[k-80])
    if k >= 120 and k <160:
        output.append(sprite[k-120])
    if k >= 160 and k <200:
        output.append(sprite[k-160])
    if k >= 200 and k <240:
        output.append(sprite[k-200])

line1 = ''.join(output[:40])
line2 = ''.join(output[40:80])
line3 = ''.join(output[80:120])
line4 = ''.join(output[120:160])
line5 = ''.join(output[160:200])
line6 = ''.join(output[200:240])
```
